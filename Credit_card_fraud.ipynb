```python
df=df.drop("Time",axis=1)
```

```python
cols = df.columns.tolist()
cols = [c for c in cols if c not in ["Class"]]
target = "Class"
```

```python
X=df[cols]
```

```python
Y=df[target]
```

```python
from imblearn.under_sampling import RandomUnderSampler 
from imblearn.over_sampling import RandomOverSampler
from imblearn.pipeline import Pipeline
ros=RandomOverSampler(sampling_strategy=0.1)
rus=RandomUnderSampler(sampling_strategy=0.5)
pipeline=Pipeline(steps=[('o',ros),('u',rus)])
x,y=pipeline.fit_resample(X,Y)
```

```python
x.shape,y.shape
```

```python
df1=pd.concat([x,y],axis=1)
```

```python
df1
```

```python
fig, axs = plt.subplots(ncols=1, figsize=(13,4.5))
sns.countplot(x="Class", data=df1, ax=axs)
fig.suptitle("Class repartition after undersampling and oversampling")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier as rfc 
from xgboost import XGBClassifier as xgbc
```

```python
models=["Logistic Regression","Support Vector Classifier","Random Forrest Classifier","XGBoost"]
```

```python
models=["Logistic Regression","Support Vector Classifier","Random Forrest Classifier","XGBoost"]
li=[]
model=lr(max_iter=2000)
log_model=model.fit(x_train,y_train)
log_pred=model.predict(x_test)
li.append(log_pred)

model=SVC(probability=True)
svc_model=model.fit(x_train,y_train)
svc_pred=model.predict(x_test)
li.append(svc_pred)

model=rfc()
rfc_model=model.fit(x_train,y_train)
rfc_pred=model.predict(x_test)
li.append(rfc_pred)

model=xgbc()
xgbc_model=model.fit(x_train,y_train)
xgbc_pred=model.predict(x_test)
li.append(xgbc_pred)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
i=0
for pred in li:
    matrix = confusion_matrix(y_test,pred)
    cm = pd.DataFrame(matrix, index=['not_fraud', 'fraud'], columns=['not_fraud', 'fraud'])

    sns.heatmap(cm, annot=True, cbar=None, cmap="Reds", fmt = 'g')
    plt.title("Confusion Matrix : "+models[i]), plt.tight_layout()
    plt.ylabel("True Class"), plt.xlabel("Predicted Class")
    plt.show()
    print("Accuracy : ",metrics.accuracy_score(y_test, pred))
    print("Precision : ",metrics.precision_score(y_test, pred))
    print("Recall : ",metrics.recall_score(y_test, pred))
    print("F1 Score : ",metrics.f1_score(y_test, pred))
    i=i+1
```
